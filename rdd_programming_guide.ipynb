```python
from pyspark import SparkContext, SparkConf
from vluigi.connections import open_connection
```

```python
conf = SparkConf().setAppName("appName").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
import pandas as pd
```

```python
data = [1, 2, 3, 4, 5]
distData = sc.parallelize(data, 10)
distData.reduce(lambda a, b: a + b)
```

```python
def myFunc(s):
    words = s.split(" ")
    return len(words)
```

```python
rdd = sc.parallelize(range(1, 4)).map(lambda x: (x, "a" * x))
rdd.collect()
```

```python
dataset = pd.read_csv("stop-word-list.csv")
data = sc.parallelize(dataset)
lineLengths = data.map(lambda s: len(s))
lineLengths.reduce(lambda a, b: a + b)
```
